```python
import cv2
import numpy as np 
import matplotlib.pyplot as plt
```

```python
blank_img = np.zeros(shape=(512, 512, 3), dtype=np.int16)
```

```python
blank_img.shape
```

```python
plt.imshow(blank_img)
```

```python
cv2.rectangle(blank_img, pt1=(300, 10), pt2=(400, 110), color=(0, 255, 0), thickness=10)
```

```python
plt.imshow(blank_img)
```

```python
cv2.circle(img=blank_img, center=(250, 250), radius=100, color=(255, 0, 0), thickness=8)
plt.imshow(blank_img)
```

```python
cv2.circle(img=blank_img, center=(410, 410), radius=50, color=(0, 0, 255), thickness=-1)
plt.imshow(blank_img)
```

```python
cv2.line(blank_img, pt1=(20, 10), pt2=(100, 400), color=(112, 255, 255), thickness=5)
plt.imshow(blank_img)
```
