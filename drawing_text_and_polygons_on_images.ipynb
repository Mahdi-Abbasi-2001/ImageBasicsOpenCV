```python
import cv2
import numpy as np 
import matplotlib.pyplot as plt
```

```python
blank_img = np.zeros(shape=(512, 512, 3), dtype=np.int16)
```

```python
cv2.putText(img=blank_img, text="Hello World!", org=(50, 200), fontFace=cv2.FONT_HERSHEY_SIMPLEX, fontScale=2,
    color=(255, 255, 255), thickness=3, lineType=cv2.LINE_AA)
plt.imshow(blank_img)
```

```python
vertices = np.array([[100, 300], [100, 400], [200, 450], [400, 350]], dtype=np.int32)
```

```python
vertices
```

```python
vertices.shape
```

```python
pts = vertices.reshape((-1, 1, 2))
```

```python
pts.shape
```

```python
pts
```

```python
cv2.polylines(blank_img, [pts], isClosed=True, color=(255, 255, 122), thickness=5)
plt.imshow(blank_img)
```
